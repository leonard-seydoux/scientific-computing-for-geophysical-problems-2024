## 2. Synthetic data

### 2.1. Synthetic monochromatic signal

In this section, we will generate a synthetic time series and explore its properties. The time series is defined as:

$$
x(t) = \sin\left(2\pi \frac{t}{t_0} + \phi\right),
$$

where $t_0$ is the period of the signal, $t$ is the continuous time, and $\phi$ is the phase of the signal. In practice, we will generate a discrete time series, where the time $t$ is sampled at discrete intervals $\delta t$. The discrete time series $x[n]$ is defined as:

$$
x[n] = \sin\left(2\pi \frac{n \delta t}{t_0} + \phi\right), \quad n = 0, 1, 2, \ldots, N-1,
$$

where $N$ is the total number of samples in the time series, and $n \delta t$ is the discrete time. For simplicity, we will continue to use the notation $x(t)$ for the discrete time series, keeping in mind that we must choose $\delta t$ carefully to avoid sampling problems.

Since we will later deal with geomagnetic data, we will generate synthetic data with equivalent units (where time is expressed in years). The time series will span from the year 1850 to 2010, and will contain a single harmonic component with a period $t_0 = 10$ years. The time series will be evenly sampled every 12<sup>th</sup> of a year (_i.e._, $\delta t = 1$ month). To generate the synthetic time series, we will use the `time_range()` and `sine()` functions from the `synthetic` Python module. These functions allow us to define a time window and create a sinusoidal time series, respectively.

> __Question 1:__ What is the maximum time step $\delta t$ that we can use to sample the synthetic time series correctly? Can you derive a general formula for the maximum time step that we can use to sample a monochromatic time series with a given period $t_0$? What effect do you expect to observe if the time step is too large?
>
> __Answer 1:__ According to the Shannon-Nyquist theorem, the maximum time step that we can use to sample the synthetic time series is half the period of the signal. In this case, the period is $t_0 = 10$ years, so the maximum time step must be strictly smaller than 5 years. The general formula is $\delta t < t_0/2$, where $t_0$ is the period of the signal. If the time step is too large, aliasing effects are expected in the time series.


```python
# Time vector of the timeseries (in decimal years)
dt = 1 / 12
times = synthetic.time_range(time_start=1850, time_end=2010, time_step=dt)

# Generate a synthetic time series with 10-year period
single_component_period = 10
single_component = synthetic.sine(times, period=single_component_period)

# Plot
fig, ax = plt.subplots()
ax.plot(times, single_component)
ax.set(
    xlabel="Time $t$ in decimal years",
    ylabel="Signal amplitude $x(t)$",
    title="Single-component signal",
)
```

$$
s(t) = \sum_{i=1}^{K} x_i(t) = \sum_{i=1}^{K} \alpha_i \sin\left( 2\pi \frac{t}{t_i} + \phi_i \right),
$$

where $K$ is the number of components, $\alpha_i$ is the amplitude, $t_i$ is the period, and $\phi_i$ is the phase of the $i$<sup>th</sup> component.

### 2.3 Spectral Analysis

We now wish to analyze the spectral content of the signals we just generated. We will start with the monochromatic signal $x(t)$, and then move on to the multi-component signal $s(t)$. We recall the definition of the Fourier transform of a signal $s(t)$ as

$$
\hat{s}(f) = \int_{-\infty}^{\infty} s(t) e^{-2\pi i f t} \, dt,
$$

where $f$ is the frequency (inverse of the period) and $\hat{s}(f)$ is the Fourier transform of $s(t)$. The _Discrete Fourier Transform_ (DFT) of a discrete time series $s[n]$ is defined as:

$$
\hat{s}[k] = \sum_{n=0}^{N-1} s[n] e^{-2\pi i k n / N},
$$

where $k$ is the frequency index and $\hat{s}[k]$ is the DFT of $s[n]$. The DFT is computed using the Fast Fourier Transform (FFT) algorithm, which is implemented in the `numpy.fft.fft` function. The FFT algorithm is very efficient and can be used to compute the DFT of a time series with a large number of samples. For simplicity, we will continue to use the notation $\hat{s}(f)$ for the DFT of $s(t)$, but remember that we must choose $\delta t$ carefully to avoid sampling issues.

Also, note that the output of a Fourier transform is a complex function. We can always express it as:

$$
\hat{s}(f) = a(f) e^{i \phi(f)},
$$

where $a(f)$ is the spectral amplitude and $\phi(f)$ is the spectral phase. Typically, we plot the spectral amplitude $a(f)$, which is a real function, as a function of the frequency $f$. The spectral amplitude is defined as the modulus of the Fourier transform, _i.e._,

$$
a(f) = \left | \hat{s}(f) \right | = \sqrt{\hat{s}(f) \hat{s}^*(f)},
$$

where $\hat{s}^*(f)$ is the complex conjugate of $\hat{s}(f)$. The square of the spectral amplitude is called the _power spectrum_ of the signal, or the _power spectral density_ (PSD).

You can run the cell below to observe the spectrum of the monochromatic signal $x(t)$ as a function of the frequency $f$. We will display the spectrum as a function of the logarithm of the frequency.


```python
# Compute the Fourier transform
n_fft = len(times)
spectrum = np.fft.rfft(single_component, n=n_fft)
spectral_amplitude = np.abs(spectrum)

# Calculate periods
frequencies = np.fft.rfftfreq(n_fft, d=dt)
frequencies[0] = np.nan
periods = 1 / frequencies  # Avoid division by zero

# Prepare the plot
fig, ax = plt.subplots(ncols=2, figsize=(8, 3))

# Plot the signal
ax[0].plot(times, single_component)
ax[0].set_xlabel("Time $t$ in years")
ax[0].set_ylabel("Signal amplitude $x(t)$")
ax[0].set_title("Single-component signal")
ax[0].grid()

# Plot the spectrum
ax[1].axvline(single_component_period, color="C1", label="True period")
ax[1].semilogx(periods, np.abs(spectrum))
ax[1].set_xlabel("Period $T$ in years")
ax[1].set_ylabel(r"Spectral amplitude $|\hat{x}(f)|$")
ax[1].set_title("Single-component signal's spectrum")
ax[1].grid()
ax[1].legend()

# Show
fig.tight_layout()
plt.show()
```

### 2.4. Filtering

In many ways we can understand the concept of filtering an input signal $x(t)$ to visualize its amplitude of a function of time at particular frequencies. For instance, thanks to the inverse transform, you can imagine designing a filter $\hat{h}(f)$ such that when applied to the Fourier transform of the input signal $\hat{x}(f)$, it will only keep the amplitude of the signal in a particular range of frequencies. The filtered signal $\hat{y}(f)$ is then given by

$$
\hat{y}(f) = \hat{x}(f) \hat{h}(f).
$$

The filtered signal in time domain is then given by the inverse Fourier transform of $\hat{y}(f)$, _i.e._, $y(t) = \mathcal{F}^{-1} \left \{ \hat{y}(f) \right \}$. We distinguish four types of filters: low-pass, high-pass, band-pass, and band-stop filters. The names of these filters are pretty self-explanatory. 

Note that in practice, not all filters are defined in the frequency domain. For different reasons, some filters can be defined in the time domain, or from recursive relations (typically, for infinite-duration impulse response filters). In this notebook, we will use the `scipy.signal` package to apply filters to our signals. This package provides a large number of filters, including Butterworth, Chebyshev, and Bessel filters. The package also provides functions to design filters, such as the `butter` function, which designs a Butterworth filter. The `scipy.signal` package also provides functions to apply filters to signals, such as the `lfilter` function, which applies a filter to a signal.

Now let us get acquainted with the Butterworth filters provided by the `scipy.signal` processing package of scientific python. The example below shows how to apply a bandpass filter, with corner periods equal to 5 and 15 years, to the example time series we have been using so far. 

#### Filter design

```python
# Periods for the bandpass filter
period_min, period_max = 5, 15

# Get the Nyquist frequency for expressing the filter in normalized units
nyquist = 1 / (2 * dt)
frequency_low = 1 / period_max / nyquist
frequency_high = 1 / period_min / nyquist

# Filter the signal
b, a = signal.butter(1, [frequency_low, frequency_high], btype="bandpass")
filtered = signal.filtfilt(b, a, single_component)

# Get spectrum
spectral_amplitude = np.abs(np.fft.rfft(single_component))
spectral_amplitude_filtered = np.abs(np.fft.rfft(filtered))
frequencies = np.fft.rfftfreq(len(single_component), dt)
frequencies[0] = np.nan
periods = 1 / frequencies

# Get filter spectrum
filter_frequencies, filter_spectrum = signal.freqz(b, a, fs=2 * nyquist)
filter_frequencies[0] = np.nan
filter_periods = 1 / filter_frequencies
filter_transmission = np.abs(filter_spectrum)

# Prepare the plot
fig, ax = plt.subplots(ncols=2, figsize=(8, 3))

# Plot the time series
ax[0].plot(times, single_component, label="$x(t)$")
ax[0].plot(times, filtered, label="$y(t)$")
ax[0].set_xlabel("Time $t$ in years")
ax[0].set_ylabel("Signal amplitude")
ax[0].set_title("Time domain")
ax[0].legend(loc="center right")
ax[0].grid()

# Plot the spectrum
ax[1].semilogx(periods, spectral_amplitude, label=r"$|\hat{x}(f)|$")
ax[1].semilogx(periods, spectral_amplitude_filtered, label=r"$|\hat{y}(f)|$")
ax[1].set_xlabel("Period $T$ in years")
ax[1].set_ylabel("Spectral amplitude")
ax[1].set_title("Spectral domain")
ax[1].legend(loc="center right")
ax[1].grid()

# Twin axes for the filter
ax_filter = ax[1].twinx()
ax_filter.semilogx(filter_periods, filter_transmission, "C2")
ax_filter.set_ylabel(r"Filter amplitude $|\hat{h}(f)|$", color="C2")
ax_filter.axvline(period_min, color="C2", linestyle="--")
ax_filter.axvline(period_max, color="C2", linestyle="--")

# Show
fig.tight_layout()
plt.show()
```

### 3.1. Loading the data

The cell below reads the file and and plots the fluctuations of the vertical component at the site between 1936 and 2015. It uses the Pandas library, which is a powerful library for data analysis. The Pandas library provides a `read_fwf` function that allows us to read a fixed-width file and store the data in a Pandas dataframe. A dataframe is a data structure that is similar to a table, where each column is a variable and each row is an observation. The Pandas library also provides a `plot` function that allows us to plot the data in a dataframe. The `plot` function is a wrapper around the `matplotlib` library, which is a powerful library for plotting data. 

```python
# Path to the data
filename = "data/chambon_la_foret.dat"

# Load the data
data = pd.read_fwf(filename, skiprows=9, strip=True)

# Turn years into datetime objects
# First, concatenate the year and month columns
date_time_concatenated = data.date.astype(str) + " " + data.time.astype(str)

# Then, convert to datetime objects
data["datetime"] = pd.to_datetime(date_time_concatenated)

# Drop the columns that are not needed anymore
data.drop(["date", "time", "doy"], axis=1, inplace=True)

# Set the datetime column as index
data.set_index("datetime", inplace=True)

# Finally, plot the vertical component of the magnetic field
ax = data.clfz.plot(grid=True, ylabel="$B_z(t)$ in nT at Chambon-la-ForÃªt")
ax.set_xlabel("Time $t$ in years")

# Show
plt.show()
```
